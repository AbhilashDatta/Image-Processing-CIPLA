```python
img_path = 'Downloads/WallTouch.png'
```

```python
import cv2
import mediapipe as mp
import imutils
import numpy as np
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
def touch(image):
    start_point = (250,0)
    end_point = (250,500)
    color = (0, 255, 0)
    thickness = 4
    stage = "Distant"
    
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image.flags.writeable = False

    # Make detection
    pose = mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5)
    results = pose.process(image)

    # Recolor back to BGR
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    # Extract landmarks
    try:
        landmarks = results.pose_landmarks.landmark

        # Get coordinates
        left_hip = [landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].x,landmarks[mp_pose.PoseLandmark.LEFT_HIP.value].y]
        right_hip = [landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_HIP.value].y]
        
        left_shoulder = [landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.LEFT_SHOULDER.value].y]
        right_shoulder = [landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].x,landmarks[mp_pose.PoseLandmark.RIGHT_SHOULDER.value].y]

        
        # Calculating Distance
        dis1 = abs(left_shoulder[0]-250)
        dis2 = abs(right_shoulder[0]-250)
        dis3 = abs(left_hip[0]-250)
        dis4 = abs(right_hip[0]-250)
        
#         print(dis1,dis2,dis3,dis4)
        if dis1<250 or dis2<250 or dis3<250 or dis4<250:
            stage = "Touching"
            
        else:
            stage = "Distant"


    except:
        pass


    # Setup status box
    cv2.rectangle(image, (0,0), (280,80), (245,117,16), -1)
    image = cv2.line(image, start_point, end_point, color, thickness)

     # Stage data
    cv2.putText(image, 'STATUS', (100,12), 
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,0), 1, cv2.LINE_AA)
    cv2.putText(image, stage, 
                (50,50), 
                cv2.FONT_HERSHEY_SIMPLEX, 1.5, (255,255,255), 2, cv2.LINE_AA)


    # Render detections
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS,
                            mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=2), 
                            mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2) 
                             )               


    cv2.imshow('Wall Touch Detector', image)
    return image 


image = cv2.imread(img_path)
image = imutils.resize(image, width=800)

image = touch(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
